```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from DecisionTreeClassifier import DecisionTreeClassifier as DTC
from sklearn.model_selection import train_test_split
```

```python
%load_ext autoreload
%autoreload 2
```

```python
data=pd.read_csv('file_path/car_evaluation.csv')
```

```python
data.columns=['buying_price','maint_price','n_doors','n_persons','lug_boot','safety','decision']
data['decision']=data['decision'].replace({'good':'acc','vgood':'acc'})
mapping={'acc':0,'unacc':1}
data['decision_encoded']=data['decision'].map(mapping)
data.head()
```

```python
X=data.drop(['decision','decision_encoded'],axis=1)
y=data['decision_encoded']
```

```python
X_train, X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
dtc_model=DTC()
```

```python
dtc_model.fit(X_train,y_train)
```

```python
y_predicted=dtc_model.predict(X_test)
```

```python
#Evaluation metrics

con_mat=dtc_model.get_confusion_matrix(y_test,y_predicted)
```

```python
accuracy=dtc_model.accuracy_score(y_test,y_predicted)
precision=dtc_model.precision_score(y_test,y_predicted)
recall=dtc_model.recall_score(y_test,y_predicted)
f1_score=dtc_model.f1_score(y_test,y_predicted)
fpr=dtc_model.fpr(y_test,y_predicted)
specificity=dtc_model.specificity(y_test,y_predicted)
```

```python
print(f'Accuarcy is {accuracy:.2f}')
print(f'Precision is {precision:.2f}')
print(f'Recall is {recall:.2f}')
print(f'F1 score is {f1_score:.2f}')
print(f'FPR is {fpr:.2f}')
print(f'Specificity is {specificity:.2f}')
```
